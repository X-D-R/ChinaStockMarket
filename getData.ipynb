```python
import yfinance as yf
generation = [str(i).zfill(6) + ".SZ" for i in range(1, 3000)]
```

```python
import pandas as pd

data = pd.DataFrame(columns=generation)

for gen in generation:
    data[gen] = yf.download(gen,'2016-01-01','2016-12-31')['Adj Close']

data.head()
```

```python
data.to_excel('output.xlsx', index=False)
```
